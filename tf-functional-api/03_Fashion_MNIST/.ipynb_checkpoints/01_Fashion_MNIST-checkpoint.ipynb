### Fashion-MNIST dataset - ``Functional API``.
In this one we are going to make use of the `Functional` api in keras to classify different clothes of 10 classes. We are going to futher on predict if the the **cloth** is esential or not.


### Imports

````python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import datasets
````

### Class Names

````python
class_names = ['T-shirt/top', 'Trouser', 'Pullover',
               'Dress', 'Coat', 'Sandal', 'Shirt', 
               'Sneaker', 'Bag', 'Ankle boot']
class_esentials =  [0, 1, 3, 6, 7]

class_names_esential = ["not esential", "esential"]
````

### Data Prep.

````python
(X_train, y_train), (X_test, y_test) = datasets.fashion_mnist.load_data()
````

````python
y_train_esential = tf.constant([1 if i in class_esentials else 0 for i in y_train])
y_test_esential = tf.constant([1 if i in class_esentials else 0 for i in y_test])

y_train_tensors = tf.one_hot(tf.convert_to_tensor(y_train), depth=10)
y_test_tensors = tf.one_hot(tf.convert_to_tensor(y_test), depth=10)

X_train_tensors = tf.convert_to_tensor(X_train)
X_test_tensors = tf.convert_to_tensor(X_test)
````

````python
print(y_train_esential[:5])
print(y_train[:5])
````

> Visualising some ``images``.

````python
def plot_images(images_and_classes, labels_true, y_test_esential, cols=7):
  rows = 3
  fig = plt.figure()
  fig.set_size_inches(cols * 2, rows * 2)
  for i, (image, label_true, j) in enumerate(zip(images_and_classes, labels_true, y_test_esential)):
    plt.subplot(rows, cols, i + 1)
    plt.axis('off')
    plt.imshow(image, cmap="gray")
    plt.title(f'{class_names[label_true]} ({j})', color ='g', fontsize=16 )

plot_images(X_test_tensors[:21], y_test[:21], y_test_esential )
````

> Createing a `Functional` model with the following `achitecture`.

```python
            input(28 x 28) image
                    |
                -----------
                | Network  |
                ------------
                |          |
            output1      output2

output1 = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress',
           'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'](Categorical)
output2 = ["essential", "not esential"] (Binary)

```

````python
input_layer = keras.layers.Input(shape=(28, 28), name="input_layer")
flatten_layer = keras.layers.Flatten(name='flatten_layer')(input_layer)
hl_1 = keras.layers.Dense(128, activation='relu', name="hidden_layer_1")(flatten_layer)
hl_2 = keras.layers.Dense(64, activation='relu', name="hidden_layer_2")(hl_1)
output_layer1 = keras.layers.Dense(10, activation='softmax', name="output_layer1")(hl_2)
output_layer2 = keras.layers.Dense(1, activation='sigmoid', name="output_layer2")(hl_2)

model = keras.Model(inputs=input_layer, outputs=[output_layer1, output_layer2],
                    name="fashion_mnist_classifier")
model.compile(
    loss = {"output_layer1": keras.losses.CategoricalCrossentropy(from_logits=False),
            "output_layer2": keras.losses.BinaryCrossentropy(from_logits=False)
           },
    optimizer ='adam',
    metrics = ['acc']
)
model.summary()
````

> Combiling the ``model``.

````python
y_tensors ={"output_layer1": y_train_tensors,
            "output_layer2": y_train_esential
            },
````

````python
X_train_tensors.shape
````

````python
EPOCHS = 10
VALID_SPLIT = .2
BATCH_SIZE = 32
VERBOSE = 1
history = model.fit(X_train_tensors, y=[y_train_tensors, y_train_esential], epochs=EPOCHS,
         verbose=VERBOSE, validation_split= VALID_SPLIT,
         batch_size=BATCH_SIZE
)
````

### Plotting the model `history`.

````python
import pandas as pd
````

````python
pd.DataFrame(history.history).plot(xlabel='epochs', title="History")
plt.show()
````

> Evaluating the `model`.

````python
model.evaluate(X_test_tensors, [y_test_tensors, y_test_esential], verbose=1)
````

> Making predictions.

````python
predictions = model.predict(X_test_tensors)
````

````python
category, is_esential = predictions
````

````python
categories_pred = tf.argmax(category, axis=1)
is_esential_pred = tf.squeeze(tf.round(is_esential))
categories_pred, is_esential_pred
````

````python
def plot_predictions_images(images_and_classes, labels_true, 
                            labels_pred, is_esential_pred,y_test_esential, cols=5):
  rows = 3
  fig = plt.figure()
  fig.set_size_inches(cols * 4, rows * 4)
  plt.title("Predictions")
  for i, (image, label_true, label_pred, label_pred_es, esent_true) in enumerate(zip(images_and_classes, 
                                                          labels_true, labels_pred, 
                                                            is_esential_pred, y_test_esential)):
    plt.subplot(rows, cols, i + 1).set_ylabel('common xlabel')
    plt.axis('off')
    plt.imshow(image, cmap='gray')
    plt.title(f'{class_names[label_true]} ({class_names_esential[label_pred_es.numpy().astype("int32")]})',
              color ='g' if label_true == label_pred and esent_true == label_pred_es.numpy().astype("int32") else 'r', fontsize=16 )

plot_predictions_images(X_test_tensors[:15], y_test[:15], 
                        categories_pred[:15], is_esential_pred[:15],
                      y_test_esential=y_test_esential.numpy())
````

> This is one of the advantages of using the `Functional` api over the ``Sequential`` api.
