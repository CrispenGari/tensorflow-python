### Transfer Learning in Keras with ``Computer Vision`` Models

A way to short-cut this process is to re-use the model weights from pre-trained models that were developed for standard computer vision benchmark datasets, such as the ImageNet image recognition tasks. Top performing models can be downloaded and used directly, or integrated into a new model for your own computer vision problems.

> âœ” Usefull Resource: [ILSVRC](https://image-net.org/challenges/LSVRC/)

### Keras Applications
[Keras Applications](https://keras.io/api/applications/) are deep learning models that are made available alongside pre-trained weights. These models can be used for prediction, feature extraction, and fine-tuning.

### Loading different models from `tf.keras.applications`

```python
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

### Classify ImageNet classes with ``ResNet50``

```python
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions, ResNet50
```

```python
model = ResNet50(weights='imagenet')
model.summary()
```

```python
image = keras.preprocessing.image.load_img('images/elephant.jpg', target_size=(224, 224))
image
```

```python
image_array = keras.preprocessing.image.img_to_array(image)
elephants = np.expand_dims(image_array, axis=0)
```

```python
elephant = preprocess_input(elephants)
```

```python
predictions = model.predict(elephant)
```

```python
decode_predictions(predictions)
```

### Extract features with ``VGG16``

```python
from tensorflow.keras.applications.vgg16 import preprocess_input, VGG16
```

```python
model = VGG16(weights='imagenet')
model.summary()
```

### Extract features from an arbitrary intermediate layer with ``VGG19``

```python
from tensorflow.keras.applications.vgg19 import preprocess_input, VGG19
from tensorflow.keras import Model
```

```python
model = VGG19(weights="imagenet")
model.summary()
```

```python
base_model = VGG19(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.get_layer('block4_pool').output)
```

```python
block4_pool_features = model.predict(elephant)
block4_pool_features
```
