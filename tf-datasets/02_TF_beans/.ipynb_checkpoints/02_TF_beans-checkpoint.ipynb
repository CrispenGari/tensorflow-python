## Beans Dataset

Beans is a dataset of images of beans taken in the field using smartphone cameras. It consists of 3 classes: 2 disease classes and the healthy class. Diseases depicted include Angular Leaf Spot and Bean Rust. 

### Imports

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow.keras as keras
import tensorflow_datasets as tfds
import pandas as pd
```

### Data Preparetion

```python
train = tfds.load('beans', split='train')
validation = tfds.load('beans', split='validation')
test = tfds.load('beans', split='test')
```

### Class Names

```python
class_names = np.array(['angular_leaf_spot', 'bean_rust', 'health'])
```

### Iterating over the datasets.

```python
train_images = tf.constant([i['image'].numpy().astype('float32')/255 for i in train])
train_labels = tf.constant([i['label'].numpy() for i in train])

test_images = tf.constant([i['image'].numpy().astype('float32')/ 255 for i in test])
test_labels = tf.constant([i['label'].numpy() for i in test])

validation_images = tf.constant([i['image'].numpy().astype('float32')/ 255 for i in validation])
validation_labels = tf.constant([i['label'].numpy() for i in validation])
```

```python
train_images[0]
```

### Visualizing images

```python
def display_images(images_and_classes, labels, cols=5):
  rows = 3
  fig = plt.figure()
  fig.set_size_inches(cols * 2, rows * 2)
  for i, (image, label) in enumerate(zip(images_and_classes, labels)):
    plt.subplot(rows, cols, i + 1)
    plt.axis('off')
    plt.imshow(image)
    plt.title(class_names[label])

display_images(test_images[:15], test_labels)
```

### Conveting Images to gray scale

```python
# X_test = test_images[:,:, :, 0]
# X_validation = validation_images[:,:, :, 0]
# X_train = train_images[:,:, :, 0]

X_test = test_images
X_validation = validation_images
X_train = train_images
```

```python
plt.imshow(X_train[0], cmap='gray')
```

### SAVING PROCESSED IMAGES

```python
train = np.save('train.npy', X_train )
test = np.save('test.npy', X_test )
validation = np.save('validation.npy', X_validation )


train_labels = np.save('train_labels.npy', train_labels )
test_labels = np.save('test_labels.npy', test_labels )
validation_labels = np.save('validation_labels.npy', validation_labels )

print("Saved")
```

### Loading the Processed Imges and labels

```python
X_train = np.load('train.npy')
X_test = np.load('test.npy')
X_validation = np.load('validation.npy')


y_train = np.load('train_labels.npy')
y_test = np.load('test_labels.npy')
y_validation = np.load('validation_labels.npy')
print("Loaded")
```

```python
X_train.shape
```

### Model Creation `CNN` Image Classifier

```python
input_shape = X_train[0].shape
input_shape
```

```python
model = keras.Sequential([
      keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
      keras.layers.MaxPool2D((2,2)),
      keras.layers.Conv2D(64, (3, 3), activation='relu'),
      keras.layers.MaxPool2D((2,2)),
      keras.layers.Conv2D(16, (3, 3), activation='relu'),
      keras.layers.MaxPool2D((2,2)),
      keras.layers.Flatten(),
      keras.layers.Dense(16, activation='relu'),
      keras.layers.Dense(128, activation='relu'),
      keras.layers.Dense(64, activation='relu'),
      keras.layers.Dense(32, activation='relu'),
      keras.layers.Dense(3, activation='softmax'),
])

model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer = keras.optimizers.Adamax(),
    metrics =['accuracy']
)
```

```python
y_train.shape
```

```python
EPOCHS = 5
BATCH_SIZE = 4
VALIDATION_SET = (X_validation, y_validation)
VERBOSE = 1

history = model.fit(X_train, y_train, epochs=EPOCHS, validation_data=VALIDATION_SET, verbose=VERBOSE, batch_size=BATCH_SIZE)
```

### Evaluating the model

```python
model.evaluate(X_test, y_test, verbose=1)
```

### Ploting the history

```python
pd.DataFrame(history.history).plot()
plt.show()
```

### Making Predictions

```python
predictions = tf.argmax(model.predict(X_test[:9]), axis=1)
```

```python
predictions 
```

```python
def display_images(images_and_classes, labels, predictions, cols=3):
  rows = 3
  fig = plt.figure()
  fig.set_size_inches(cols * 5, rows * 5)
  for i, (image, label, pred) in enumerate(zip(images_and_classes, labels, predictions)):
    plt.subplot(rows, cols, i + 1)
    plt.axis('off')
    plt.imshow(image)
    plt.title(f'True: {class_names[label]} Pred: {class_names[pred]}')
display_images(X_test[:9], y_test, predictions)
```
